```python
rows = [[]]
genres = list(uniqueGenres)

for i in data.genres:
  temp = ast.literal_eval(i)
  temp2 = [x['name'] for x in temp]
  row = []
  for i in genres:
    row.append(int(i in temp2))
  rows.append(row)
```
